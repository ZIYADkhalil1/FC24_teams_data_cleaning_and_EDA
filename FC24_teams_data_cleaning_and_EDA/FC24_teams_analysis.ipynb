let's fix the home stadim

```python
null_rows = fc24[fc24['home_stadium'].isnull()]

pd.DataFrame(null_rows)
```

those teams home stadium isnt included in FC24 , but we'll replace it with thier actual home stadium name

```python
pd.DataFrame(null_rows)['team_name'].unique()
```

we'll create a dictionary that has each team name and stadium name, then it iterates through the specified teams with null values in the 'home_stadium' column and replaces the null values with the corresponding home stadium names from the team_stadium_mapping dictionary

```python
team_stadium_mapping = {
    'Atalanta': 'Stadio Atleti Azzurri d\'Italia',
    'Lazio': 'Stadio Olimpico',
    'Napoli': 'Stadio San Paolo',
    'Roma': 'Stadio Olimpico',
    'Spezia': 'Stadio Alberto Picco',
    'CFR Cluj': 'Dr. Constantin RÄƒdulescu Stadium'
}

teams_with_nulls = ['Atalanta', 'Lazio', 'Napoli', 'Roma', 'Spezia', 'CFR Cluj']

for team in teams_with_nulls:
    fc24.loc[fc24['team_name'] == team, 'home_stadium'] = team_stadium_mapping.get(team, None)
```
